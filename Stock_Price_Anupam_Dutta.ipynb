```python
#Training the Model
# 1. RandomForestClassifier (RandomForestClassifier Works by training a bunch of individual decision trees with randomized parameters and then averaging the results from those desicion trees so because of this process this algo is resistent to overfitt )
# The Random Forest Classifier pick up non-linear tendencies 
# The relationship between open price and target is nonlinear
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
# n_estimators is the number of Decision trees we want to train
# n_estimators higher the number means higher accuracy to a certain limit
# min_samples_split this helps us to protect against overfitting as decision trees has a tendency to overfit if they build the tree too deeply
# min_samples_split the higher we set it the model will less accurate but the less will cause overfitting the optimal value should be searched
```

```python
#Spliting the Dataset between Train and Test Set
""" The Stocks Data is time series Data in
this type of data we cannot use the cross validation, if we use
Cross Validation in this data the results will maybe correct while training but,
wrong in real word as, if we use Cross validation or another way to split up the train and test set
that doesn't take time series nature of the data into the account & in that case we will be using the future data to predict the past which
we can't do in real world and the result of it will be the informatin leakage into the model
"""
train = sp500.iloc[:-100] # All Rows without last 100 rows in Train Set
test = sp500.iloc[-100:] # Last 100 rows in test set

predictors = ["Close", "Volume", "Open", "High", "Low"] # That all mentiond colums in the predictors list going to use to predict the target

model.fit(train[predictors],train["Target"]) #Train The Model where using the predictors columns in order to predict the target
```

```python
# Calculating the Accuracy of the model

from sklearn.metrics import precision_score
preds = model.predict(test[predictors]) # Generates the Predictions and returns in numpy array
```

```python
# Backtesting System
```

```python
# Currently we are testing our model on last 100 days
# But, in case of real word senario we must test across multiple years of data
# As, we have to understand how our algo is handeling lots of different situations

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
# Backtest Function
# Every Trading year has 250 trading days
# Here we are taking 10 years of trading data so 10 * 250 is 2500 that is why the start value is 2500
# So, the first model will be trained with 2500 days (10 years of data)
# step is 250 means that we will be training a model for about a year (1 treading year has 250 days) and then going to the next year and then going to the next year
# In more simpler way, we are going to take the first 10 years of data and then predict the value for 11th Year, After that we will take 11 years of data & predict the values for 12th year and so on.
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = [] # List of Dataframes where each data frame is the predictions for a single year
    
    for i in range(start, data.shape[0], step): # Loop across our data year by year and makes the predictions for all the years except the first 10 years
        train = data.iloc[0:i].copy() # Training set, The training set is all of the years prior to the current year and the test set is the current year
        test = data.iloc[i:(i+step)].copy() # the test set is the current year
        predictions = predict(train, test, predictors, model) # predict function to generate our predictions
        all_predictions.append(predictions) # predictions are append to all_predictions for a given year
    return pd.concat(all_predictions)
        
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
# Adding some additional Predictors to the Model (for improving the accuracy)
```

```python
# Creating some verity of rolling averages
# Checking something like is the stock price today is higher or lower then it was last week or higher or lower then it was three months ago or a year or 5 years ago, thats helps to determine if the stock will go up or down
# We are giving the algorithm of that information.

horizons = [2,5,60,250,1000] # rolling means 
# Calculate the mean close price in the last 2 days, last trading weeks (5 days), last 3 months ago (60 days), last year(250 days) & last 4 years (1000 days)
# Then we will find the ratio  between  today's closing price and the cloaseing price in those preiods which will help us to know is the market gone up because if so then it may be due for a downturn, has the market gone down if so then it maybe due for  an upswing
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"]/rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
# Currently we are testing our model on last 100 days
# But, in case of real word senario we must test across multiple years of data
# As, we have to understand how our algo is handeling lots of different situations

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] # model.predict_proba it will return a probability that the row will be 0 or 1
    preds[preds >= .6] = 1 # if the preds >= .6 then 1 will be assigend and if preds < .6 then 0 will be assign
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```
